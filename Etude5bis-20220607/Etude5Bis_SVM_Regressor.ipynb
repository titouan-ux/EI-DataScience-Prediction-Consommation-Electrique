### SVM   Regressor

1) Importer les librairies classiques : numpy, pandas, sklearn.svm, seaborn et matplolib

```python
import numpy as np
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
```

2) Lire le fichier sinusPoint.csv avec pandas

```python
df=pd.read_csv('sinusPoint.csv',sep=";")
df
```

3) Tracer y(x), sous forme de nuage de points

```python
#Tracer de la figure
plt.plot(df['x'],df['y'],'.')
plt.show()
```

4) Entrainement, test avec l'algorithme SVR

```python
from sklearn.svm import SVR


model_SVR=SVR()

#Données d'entrainement
x_train=df['x'].values.reshape(-1,1)
y_train=df['y'].values.reshape(-1,1)

#Entrainement
model_SVR.fit(x_train,y_train.ravel())

T=np.linspace(0,5,100)

#prédiction
Y_predict=model_SVR.predict(T.reshape(-1,1))

#Affichage de la prediction du modèle SVR
plt.plot(df['x'],df['y'],'r.',label='data' )
plt.plot(T,Y_predict,'k',label='SVR')
plt.title('Tracé du modèle avec SVR')



from sklearn.neighbors import  KNeighborsRegressor
neigh=KNeighborsRegressor(n_neighbors=5, weights="uniform")

#Entrainement
neigh.fit(x_train,y_train.ravel())

#Prediction
Y_predict=neigh.predict(T.reshape(-1,1))

plt.plot(T,Y_predict,'b',label='KNN regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# On compare 2 algorithmes de regression SVR et KNN et on voit que ici pour des données très disparates, SVR est vraiment plus performant
```
