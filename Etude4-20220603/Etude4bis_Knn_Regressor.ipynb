# Nearest Neighbors regression

Démontrer la résolution d'un problème de régression
en utilisant un k-Nearest Neighbor 

1) Creer une base de données. Expliquer son fonctionnement

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors

n=100
np.random.seed(0)
X = np.sort(5 * np.random.rand(n, 1), axis=0)
T = np.linspace(0, 5, 10000)[:, np.newaxis]
y = np.sin(X).ravel()

# Add noise to targets
y[::5] += 1 * (0.5 - np.random.rand(n//5))
plt.scatter(X, y, color="darkorange", label="data")
plt.legend()
print(len(X))
```

2) Créer un modèle en utilisant KNN_Regressor
<  neighbors.KNeighborsRegressor(n_neighbors,weights="uniform")  >

```python
from sklearn.neighbors import  KNeighborsRegressor

#On pose le modèle KNN pour 5 voisins
neigh=KNeighborsRegressor(n_neighbors=5, weights="uniform")
#on entraîne le modele
neigh.fit(X,y)
#On applique celui-ci a l'interval [0,5]
Y_predict=neigh.predict(T)

#On représente nos résultats sur un graphique
plt.plot(T,Y_predict,'k',label='predicted data')
plt.plot([2],neigh.predict([[2]]),'g.')
plt.plot(X, y,"b.",label='data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()


```
